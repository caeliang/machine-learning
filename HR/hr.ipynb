```python
import numpy as np  
import pandas as pd
from sklearn.model_selection import train_test_split
train=pd.read_csv('aug_train.csv')
x_train=pd.DataFrame(train.drop(['target'],axis=1))
y_train=train['target']


x_train['gender'].fillna(x_train['gender'].mode()[0], inplace=True)

x_train['enrolled_university'].fillna('unknown', inplace=True)
x_train['education_level'].fillna('unknown', inplace=True)

x_train['major_discipline'].fillna('unknown', inplace=True)
x_train['experience'].fillna('0', inplace=True)

x_train['company_size'].fillna('unknown', inplace=True)
x_train['company_type'].fillna('unknown', inplace=True)
x_train['last_new_job'].fillna('0', inplace=True)
x_train['last_new_job'] = x_train['last_new_job'].replace('>4', 5).replace('never', 0).astype(float)


x_train.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

# Deneyim sütunundaki '>' ve '<' karakterlerini kaldırın ve int türüne dönüştürün
x_train['experience'] = x_train['experience'].str.replace('>', '', regex=False).str.replace('<', '', regex=False)
x_train['experience'] = x_train['experience'].astype(int)

# Kategorik özellikleri tanımlayın
categorical_features = ['gender', 'relevent_experience', 'enrolled_university', 
                        'education_level', 'major_discipline', 'company_size', 'company_type']

# LabelEncoder örneğini oluşturun
encoder = LabelEncoder()

# Her bir kategorik özelliği dönüştürün
for feature in categorical_features:
    x_train[feature] = encoder.fit_transform(x_train[feature])

# İlk birkaç satırı görüntüleyin
x_train.head()
```

```python
x_train['experience'] = x_train['experience'].str.replace('>', '').str.replace('<', '')
x_train['experience'] = x_train['experience'].astype(int)
x_train.head()
```

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential()
model.add(Dense(12, input_shape=(8,), activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=150, batch_size=10, verbose=0)
_, accuracy = model.evaluate(x_train, y_train)
print('Accuracy: %.2f' % (accuracy*100))
```
